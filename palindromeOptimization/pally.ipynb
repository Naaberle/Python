```python
pallys = []

for x in range(1,1000):
    for y in range(1,1000):
        result = x*y
        newstr = str(result)
        if newstr == newstr[::-1]:
            pallys.append(result)
            
max(pallys)
```

```python
pallys = []

for x in range(1,1000):
    pallys.append([x*y for y in range(1,1000) if str(x*y) == str(x*y)[::-1] and ((x*y)%11 != 0)])

print(pallys)
```

```python
highest = 0

for x in range(1000,915,-1):
    for y in range(1000,912,-1):
        newhighest = x*y
        newstr = str(newhighest)
        if newstr == newstr[::-1] and newhighest > highest:
            highest = newhighest
            
print(highest)
```

```python
highest = 0

newhighest = [x*y for x in range(1000,1,-1) for y in range(1000,1,-1) if str(x*y) == str(x*y)[::-1]]

for i in newhighest:
    if i > highest:
        highest = i
        
print(highest)
```
